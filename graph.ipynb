```python
import re
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.set_style("whitegrid")
```

```python
def get_logs(filename):
    with open(filename, 'r') as file:
        lines = [line.strip() for line in file]
    train_loss = []
    train_acc = []
    val_loss = []
    val_acc = []
    for i in range(len(lines)):
        logs = re.findall(r'\d+\.\d+', lines[i])
        if i%2==0: # train
            train_loss.append(float(logs[0]))
            train_acc.append(float(logs[1]))
        else:
            val_loss.append(float(logs[0]))
            val_acc.append(float(logs[1]))
    return train_loss, train_acc, val_loss, val_acc
```

```python
train_loss, train_acc, val_loss, val_acc = get_logs('m1_out.txt')
x = [i for i in range(97)]
plt.plot(x, train_acc, label='train', color='royalblue')
plt.plot(x, val_acc, label='val', color='orange')
plt.legend(loc='upper left')
plt.savefig('m1.png', dpi=300)
```
